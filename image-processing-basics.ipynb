```python
import cv2
from matplotlib import pyplot as plt
import pytesseract
from pytesseract import Output
```

## Image en niveau de gris

```python
imagePath = r'image1.jpg'
image = cv2.imread(imagePath)
plt.imshow(image)
```

## Seuillage de l'image

Explication des options de seuillage (https://docs.opencv.org/master/d7/d1b/group__imgproc__misc.html#ggaa9e58d2860d4afa658ef70a9b1115576ac7e89a5e95490116e7d2082b3096b2b8)  
Seuil à la moitié de la colorimétrie (255), soit environ 128 :

### Seuillage classique

```python
ret,seuil1 = cv2.threshold(image,128,255,cv2.THRESH_BINARY)
plt.imshow(seuil1)
```

### Seuillage inverse

```python
seuil1
```

```python
ret,seuil1 = cv2.threshold(image,128,255,cv2.THRESH_BINARY_INV)
plt.imshow(seuil1)
```

### Seuillage tronqué

```python
ret,seuil1 = cv2.threshold(image,128,255,cv2.THRESH_TRUNC)
plt.imshow(seuil1)
```

### Seuillage à zéro à partir du seuil / on conserve le reste de l'image tel quel

```python
ret,seuil1 = cv2.threshold(image,128,255,cv2.THRESH_TOZERO)
plt.imshow(seuil1)
```

### Seuillage à 255 à partir du seuil / on conserve le reste de l'image tel quel

```python
ret,seuil1 = cv2.threshold(image,128,255,cv2.THRESH_TOZERO_INV )
plt.imshow(seuil1)
```

## Du seuillage plus fin

```python
imagePath = r'facture.png'
plt.rcParams['figure.figsize'] = [15, 15]
image = cv2.imread(imagePath)
plt.imshow(image)
```

### Niveaux de gris

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("gris", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Seuillage

```python
ret,seuil1 = cv2.threshold(image,150,255,cv2.THRESH_BINARY)
plt.imshow(seuil1)
```

```python
src = cv2.imread(imagePath, cv2.CV_8UC1);
image_blur = cv2.medianBlur(src,1)
plt.imshow(image_blur)
```

```python
seuil2 = cv2.adaptiveThreshold(image_blur,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
plt.imshow(seuil2)
```

```python
seuil3 = cv2.adaptiveThreshold(image_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
plt.imshow(seuil3)
```
